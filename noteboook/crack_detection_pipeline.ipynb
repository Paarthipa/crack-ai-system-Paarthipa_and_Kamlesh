# 🏗️ AI-Based Bridge & Road Crack Detection System

## 📌 Overview
This notebook is part of a multi-component crack detection system designed for infrastructure monitoring.

The system integrates:
- Computer Vision (CV) for crack detection
- Feature extraction (length, width, severity)
- Retrieval-Augmented Generation (RAG)
- AI-based reasoning and reporting

---

## 🎯 Objective
To detect cracks in road/bridge images and provide:


- Visual outputs (mask, overlay)
- Quantitative analysis
- AI-based insights

## 📦 Import Required Libraries

```python
import requests
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from PIL import Image
from IPython.display import display
```

## 🌐 API Configuration

We connect to the backend FastAPI server for crack detection and AI responses.  

```python
API_URL = "http://127.0.0.1:8000/analyze-crack"
CHAT_API = "http://127.0.0.1:8000/chat"
```

## 🖼️ Load and Display Test Image

We load a sample crack image for analysis.

```python
image_path = "test-case-images/crack1.jpg"

img = Image.open(image_path)
display(img)
```

## 🔍 Crack Detection

The image is sent to the API for crack analysis.


```python
with open(image_path, "rb") as f:
    response = requests.post(
        API_URL,
        files={"file": f}
    )

data = response.json()
data
```

## 📏 Crack Measurements

The system extracts crack properties.


```python
print("📏 Length:", data["crack_length"], "m")
print("📐 Width:", data["crack_width"], "mm")
print("⚠️ Severity:", data["severity"])
print("📍 Location:", data["location"])
```

## 📊 Crack Dimension Visualization


```python
fig = px.bar(
    x=["Length", "Width"],
    y=[data["crack_length"], data["crack_width"]],
    title="Crack Dimensions",
    text_auto=True
)

fig.show()
```

## ⚠️ Severity Indicator



```python
severity_map = {"Low": 1, "Medium": 2, "High": 3}
value = severity_map[data["severity"]]

fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=value,
    title={'text': "Severity"},
    gauge={'axis': {'range': [0, 3]}}
))

fig.show()
```

## 🥧 Crack Distribution


```python
fig = px.pie(
    names=["Crack", "Remaining"],
    values=[data["crack_width"], max(20 - data["crack_width"], 1)],
    hole=0.5
)

fig.show()
```

## 📈 Crack Growth Prediction


We simulate how the crack width may grow over time.

```python
time = np.arange(0, 10)
growth = data["crack_width"] * (1 + 0.1 * time)

fig = px.line(
    x=time,
    y=growth,
    title="Crack Growth Prediction"
)

fig.show()
```

## 🕸️ Multi-Factor Analysis



```python
fig = go.Figure()

fig.add_trace(go.Scatterpolar(
    r=[
        data["crack_length"],
        data["crack_width"],
        {"Low":1,"Medium":2,"High":3}[data["severity"]]
    ],
    theta=["Length", "Width", "Severity"],
    fill='toself'
))

fig.show()
```

## 🔥 Damage Heatmap

```python
heat_data = np.random.rand(10, 10) * data["crack_width"]

fig = px.imshow(heat_data, color_continuous_scale="Reds")
fig.show()
```

## 🤖 AI Assistant (Auto Query)

```python
question = "Is this crack dangerous?"

res = requests.post(
    CHAT_API,
    json={
        "question": question,
        "data": data
    }
)

print("🤖 AI:", res.json()["answer"])
```

## 💬 Ask Your Own Question


```python
user_q = input("Ask a question: ")

res = requests.post(
    CHAT_API,
    json={
        "question": user_q,
        "data": data
    }
)

print("🤖 AI:", res.json()["answer"])
```
